```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("C://Users/sampa/ai_class/data_files\AAME.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df = df.drop(columns=['Date', 'Adj Close', 'Volume'])
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0, 1))
scaled = scaler.fit_transform(df)
scaled
```

```python
int(len(scaled) * 0.8)
```

```python
int(len(scaled)) - int(len(scaled) * 0.8)
```

```python
len(scaled) == int(len(scaled) * 0.8) + int(len(scaled)) - int(len(scaled) * 0.8)
```

```python
TEST_SIZE = 2020

train_data = scaled[:-TEST_SIZE]
test_data = scaled[-TEST_SIZE:]

print(len(train_data))
print(len(test_data))
```

```python
def make_sample(data, window):
    train = []
    target = []
    for i in range(len(data)-window):
        train.append(data[i:i+window])
        target.append(data[i+window][3])
    return np.array(train), np.array(target).reshape(-1, 1)

X_train, y_train = make_sample(train_data, 30)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
model = Sequential()
model.add(SimpleRNN(16,
               input_shape=(X_train.shape[1], 4),
               activation='tanh',
         ))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=16)
```

```python
X_test, y_test = make_sample(test_data, 30)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, pred)
print(mse)
```

```python
plt.figure(figsize=(12, 9))
plt.plot(y_test, label='stock price', alpha=0.7, color='gray')
plt.plot(pred, label='predicted stock price by Simple RNN', color='red')
plt.legend()
plt.tight_layout()
plt.grid()
plt.show()
```
