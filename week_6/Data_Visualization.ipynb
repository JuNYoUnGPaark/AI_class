## **Matplotlib**
- Python의 데이터 시각화 라이브러리
- `pyplot` 모듈을 이용한 *Scripting* 방식과 `Figure` & `Axes`를 직접 정의해서 사용하는 *OOP 방식* 으로 사용 가능

```python
'''기본 사용법'''

# 라이브러리 불러오기
import matplotlib.pyplot as plt

# 데이터 정의
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# 플로팅 (그래프 그리기)
plt.plot(x, y)
plt.title("Basic Line Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.show()
```

```python
'''OOP Version'''

# 라이브러리 불러오기
import matplotlib.pyplot as plt

# 데이터 정의
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# 직접 figure와 axes를 생성한다
# Figure: 플로팅을 하는 영역 전체
# Axes: 하나의 플롯
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("OOP Style Basic Line Graph")
ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")
ax.grid(True)
plt.show()
```

```python
''' OOP 방식으로 여러 그래프 그리기 Subplot '''
''' 위치 인덱싱 방식으로 각 위치를 지정해서 그릴 수 있다. '''
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(x, y)
axs[0, 0].set_title("Plot 1")

axs[0, 1].bar(['A', 'B', 'C'], [5,7,3])
axs[0, 1].set_title("Bar")

axs[1, 0].scatter(x, y)
axs[1, 0].set_title("Scatter")

axs[1, 1].hist(np.random.randn(1000))
axs[1, 1].set_title("Histogram")

plt.tight_layout()
plt.show()
```

### **스타일링 주요 옵션**
- `color` : 선 색상
  - 이름, 약어, RGB 튜플, HEX 코드
- `linestyle` : 선 스타일
  - `-`, `:`, `--`, `-.` 
- `linewidth` : 선 두께
  - `int` 
- `marker` : 점 모양
  - `o`, `s`, `^`, `v`, `D`
- `markersize` : 점 크기
    - `int`
- `plt.grid(True)`
- `plt.xlim()`, `plt.ylim()`
- `plt.xticks()`, `plt.yticks()`
- `plt.axhline()`
- `plt.axvline()`

```python
plt.plot(x, y, color='red', linestyle='--', linewidth=2, marker='o', markersize=8)
plt.title("Custom Line Graph")
plt.show()
```

```python
plt.plot(x, y)
plt.grid(True)
plt.xlim(0, 6)
plt.ylim(0, 30)
plt.axhline(y=10, color='gray', linestyle='--')
plt.axvline(x=3, color='gray', linestyle='--')
plt.show()
```

## **Seaborn**
- Python의 데이터 시각화 전문 라이브러리
- 내부적으로 **Matplotlib 기반**
  - 더 간결한 코드
  - 더 세련된 그래프
  - pandas의 DF와 자연스러운 호환

```python
''' Basic plotting '''
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터
tips = sns.load_dataset("tips")  # seaborn 내장 데이터셋

# 시각화
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
```

### **주요 차트 종류별 예제**

```python
''' lineplot '''
sns.lineplot(data=tips, x="size", y="tip")
plt.title("Lineplot Example")
plt.show()
```

```python
''' barplot '''
sns.barplot(data=tips, x="day", y="total_bill", errorbar='sd')  # sd: 신뢰구간 표시
plt.title("Barplot Example")
plt.show()
```

```python
''' scatter plot '''
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="time")  # hue: 색상 그룹핑,  style: 마커 모양 구분
plt.title("Scatterplot Example")
plt.show()
```

```python
''' countplot '''
sns.countplot(data=tips, x="day")
plt.title("Countplot Example")
plt.show()
```

```python
''' boxplot '''
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Boxplot Example")
plt.show()
```

```python
''' violinplot '''
sns.violinplot(data=tips, x="day", y="total_bill", inner="quartile")
plt.title("Violinplot Example")
plt.show()
```

```python
''' heatmap '''

tips_corr = tips.drop(columns=["sex", "smoker", "day", "time"])

corr = tips_corr.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Heatmap Example")
plt.show()
```

```python
''' 
숫자형만 들어갈 수 있는 heatmap 사용 대신
swarmplot: 겹치지 않게 퍼뜨린 점 그래프
stripplot: 문자열 축에도 가능 (수평 산점도)
등을 사용 가능하다
'''

sns.stripplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.title("Stripplot Example")
plt.show()

sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.title("Swarmplot Example")
plt.show()
```

```python
''' pairplot: 변수 간 관계를 일괄 비교 '''
sns.pairplot(tips, hue="sex")
plt.title("Pairplot Example")  # 자동으로 여러 그래프 그려짐
plt.show()
```

```python
''' jointplot '''
sns.jointplot(data=tips, x="total_bill", y="tip", kind="hex")
plt.show()
```

### **데이터 시각화 분석 예시 by titanic.csv**

```python
# 데이터 불러오기
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Titanic 데이터셋 로드
titanic = sns.load_dataset('titanic')

# 데이터 확인
print(titanic.head())
```

```python
# 생존 여부 비율
sns.countplot(data=titanic, x='survived')
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# 성별별 생존 비율
sns.countplot(data=titanic, x='sex', hue='survived')
plt.title('Survival by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
# 나이 분포
sns.histplot(data=titanic, x='age', kde=True)  # kde: 밀도 곡선
plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()
```

```python
# 나이와 생존 관계
sns.boxplot(data=titanic, x='survived', y='age')
plt.title('Age vs Survived')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

```python
# 클래스별 생존율
sns.countplot(data=titanic, x='pclass', hue='survived')
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
# 등급, 성별, 생존율 복합 분석
g = sns.FacetGrid(titanic, col="sex", row="pclass", margin_titles=True)
g.map_dataframe(sns.histplot, x="age", hue="survived", multiple="stack")
g.add_legend()
plt.show()
```

```python
# 요금과 나이 관계
sns.scatterplot(data=titanic, x='age', y='fare', hue='survived')
plt.title('Age vs Fare (colored by Survival)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# 변수별 상관관계
corr = titanic[['age', 'fare', 'pclass', 'survived']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(titanic[['age', 'fare', 'pclass', 'survived']], hue='survived')
plt.suptitle('Pairplot of Age, Fare, Pclass', y=1.02)
plt.show()
```

```python
pivot = titanic.pivot_table(index='sex', columns='pclass', values='survived')
sns.heatmap(pivot, annot=True, cmap='YlGnBu')
plt.title('Survival Rate by Sex and Class')
plt.show()
```
